```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

# Update file path
file_path = '/content/drive/MyDrive/zinoanalytics-portfolio/naira-vs-dollar/naira_usd_2008:2025.csv'

import pandas as pd
df = pd.read_csv(file_path)

df.head()

# Keep only Date and Value columns
df = df[['Date', 'Value']].copy()
df.rename(columns={'Value': 'Exchange Rate'}, inplace=True)

# Clean Date column
df['Date'] = pd.to_datetime(
    df['Date'].str.replace('M', '-', regex=False).str.replace(r'-(\d)$', r'-0\1', regex=True),
    format='%Y-%m'
)

# Sort by date
df = df.sort_values('Date')

# Preview cleaned data
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set seaborn style
sns.set(style='whitegrid')

# Create line plot
plt.figure(figsize=(14,6))
sns.lineplot(data=df, x='Date', y='Exchange Rate', linewidth=2.2, color='darkblue')

# Chart formatting
plt.title('Naira to Dollar Exchange Rate (2008–2025)', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Exchange Rate (₦ per $1)')
plt.tight_layout()
plt.show()
```

### 🧠 Key Insights:
- The Naira has consistently depreciated against the US Dollar from 2008 to 2025.
- Sharp increases are visible around 2016, 2020, and mid-2023 — likely tied to recession, COVID-19, and FX policy reforms.
- Despite brief periods of stability, the long-term trend shows continuous pressure on the currency.
